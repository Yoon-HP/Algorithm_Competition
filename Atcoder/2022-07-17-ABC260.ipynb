```python
# B-Better Students Are Needed! (성공)
import heapq
N,X,Y,Z=map(int,input().split())
ans=[]
M=[0]+list(map(int,input().split()))
E=[0]+list(map(int,input().split()))
temp_M=[]
temp_E=[]
total=[]
for i in range(1,N+1):
    heapq.heappush(temp_M,[-M[i],i])
    heapq.heappush(temp_E,[-E[i],i])
    heapq.heappush(total,[-M[i]-E[i],i])
check=[False]*(N+1)

for i in range(X):
    w,n=heapq.heappop(temp_M)
    check[n]=True
    ans.append(n)
    
for i in range(Y):
    while True:
        w,n=heapq.heappop(temp_E)
        if not check[n]:
            check[n]=True
            ans.append(n)
            break
for i in range(Z):
    while True:
        w,n=heapq.heappop(total)
        if not check[n]:
            check[n]=True
            ans.append(n)
            break
ans.sort()
for i in range(len(ans)):
    print(ans[i])
```

```python
# C- Changing Jewels (성공)

N,X,Y=map(int,input().split())
if N==1:
    print(0)
else:
    blue=[-1]*(N+1)
    blue[2]=Y
    def jewel(N):
        if blue[N]!=-1:
            return blue[N]
        if N>=2:
            temp=0
            for i in range(2,N):
                temp+=X*jewel(i)
            blue[N]=temp+Y*jewel(N-1)
        return blue[N]
    ans=0
    jewel(N)
    for i in range(2,N+1):
        ans+=blue[i]*X
    print(ans)
```

```python
# D-Draw Your Cards (성공)
import bisect

N,K=map(int,input().split())
ans=[-1]*(N+1)
deck=list(map(int,input().split()))
if K==1:
    for i in range(N):
        ans[deck[i]]=i+1
else:
    h=[]
    ref=[]
    cnt=0
    current_max=0
    # i+1은 라운드
    for i in range(N):
            if not ref:
                ref.append(deck[i])
                # 해당되는 수, 해당되는 수의 개수
                h.append([[deck[i]],1])
                cnt+=1
            else:
                p=bisect.bisect_left(ref,deck[i])
                if p==cnt:
                    ref.append(deck[i])
                    h.append([[deck[i]],1])
                    cnt+=1
                else:
                    ref[p]=deck[i]
                    h[p][0].append(deck[i])
                    h[p][1]+=1
                    if h[p][1]==K:
                        for j in h[p][0]:
                            ans[j]=i+1
                        cnt-=1
                        ref.remove(deck[i])
                        h.remove(h[p])
        
for i in range(1,N+1):
    print(ans[i])
```
