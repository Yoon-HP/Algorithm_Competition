```python
# A. Mark the Photographer

t=int(input())
for _ in range(t):
    n,x=map(int,input().split())
    first=[]
    h=list(map(int,input().split()))
    h.sort()
    f_i=0
    check=True
    for i in range(n):
        first.append(h[i])
    for i in range(n,2*n):
        if h[i]-first[f_i]>=x:
            f_i+=1
        else:
            check=False
    if check:
        print("YES")
    else:
        print("NO")
```

```python
# B. Mark the Dust Sweeper

t=int(input())

for _ in range(t):
    n=int(input())
    room=list(map(int,input().split()))
    ans=0
    ck=0
    for i in range(n-1):
        if ck==i and room[i]==0:
            ck+=1
            continue
        if room[i]!=0:
            ans+=room[i]
        else:
            ans+=1
    print(ans)
```

```python
# C. Mark and His unfinished Essay

t=int(input())
for _ in range(t):
    n,c,q=map(int,input().split())
    s=input().strip()
    paste=[]
    paste.append([1,n,n])
    for i in range(1,c+1):
        l,r=map(int,input().split())
        paste.append([l,r,paste[i-1][2]+(r-l+1)])
    # k는 인덱스
    for i in range(q):
        k=int(input())
        while True:
            if k<=n:
                break
            for j in range(c+1):
                if k<=paste[j][2]:
                    k=k-paste[j-1][2]+paste[j][0]-1
                    break
        print(s[k-1])
```
