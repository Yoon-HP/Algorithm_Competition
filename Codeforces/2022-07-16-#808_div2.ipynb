```python
# A. Difference Operations (성공)
# 문제의도 a2,a3,a4,...,an까지 각 수가 a1의 배수인지 아닌지 확인하는 것 모두 a1의 배수라면 YES
# 뒤집어서 푸는 것도 가능
t=int(input())
for _ in range(t):
    n=int(input())
    arr=list(map(int,input().split()))
    for i in range(1,n):
        if arr[i]>arr[i-1]:
            arr[i]=arr[i]-(arr[i]//arr[i-1])*arr[i-1]
            if arr[i]==0:
                arr[i]=arr[i-1]
    arr=arr[::-1]
    check=True
    for i in range(n-1):
        if arr[i+1]>arr[i]:
            check=False
            break
        if arr[i]%arr[i+1]!=0:
            check=False
            break
    if check:
        print("YES")
    else:
        print("NO")
```

```python
# B. Difference of GCDs (성공)
# gcd(i,ai)=i라는 것을 파악하는게 문제 의도.
t=int(input())
for _ in range(t):
    n,l,r=map(int,input().split())
    ans=[]
    check=True
    for i in range(1,n+1):
        if i>l:
            if i<=r:
                ans.append(i)
            else:
                check=False
                break
        else:
            temp=l%i
            if temp==0:
                ans.append(l)
            else:
                if l+(i-temp)<=r:
                    ans.append(l+(i-temp))
                else:
                    check=False
                    break
    if check:
        print("YES")
        print(*ans)
    else:
        print("NO")
```

```python
# C. Doremy's IQ (실패)
import heapq

t=int(input())
for _ in range(t):
    n,q=map(int,input().split())
    
    if q>=n:
        s="1"*n
        print(s)
        input()
    else:
        cnt=0
        arr=list(map(int,input().split()))
        h=[[-n+q,q,0,'']]
        check=False
        while h:
            print(h)
            w,cur_q,index,temp=heapq.heappop(h)
            if index==n:
                print(temp)
                break
            if cur_q==0:
                continue
            if cur_q>=n-index:
                check=True
                break
            if arr[index]<=cur_q:
                heapq.heappush(h,[w+1,cur_q,index+1,temp+'1'])
            else:
                # 참여 하는 경우
                heapq.heappush(h,[w,cur_q-1,index+1,temp+'1'])
                # 참여 안하는 경우
                heapq.heappush(h,[w+1,cur_q,index+1,temp+'0'])
        if check:
            print(temp+"1"*cur_q)
```

```python
# 업솔빙
# C. Doremy's IQ

t=int(input())
for _ in range(t):
    n,q=map(int,input().split())
    Q=0
    ans=''
    arr=list(map(int,input().split()))
    for i in range(-1,-n-1,-1):
        if arr[i]>Q:
            if Q<q:
                ans+='1'
                Q+=1
            else:
                ans+='0'
        else:
            ans+='1'
    print(ans[::-1])
```
