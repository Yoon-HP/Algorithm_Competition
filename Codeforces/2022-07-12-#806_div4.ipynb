```python
# C Cypher

t=int(input())
for _ in range(t):
    n=int(input())
    current=list(map(int,input().split()))
    for i in range(n):
        b,s=input().split()
        temp=s.count("D")
        change=abs(int(b)-temp-temp)
        if (int(b)-temp>temp):
            if change>current[i]:
                current[i]=10-(change-current[i])
            else:
                current[i]=current[i]-change
        else:
            if (change+current[i]<=9):
                current[i]=current[i]+change
            else:
                current[i]=change+current[i]-10
    for i in current:
        print(i,end=" ")
    print()
```

```python
# D Double Strings
import sys
class Node(object):
    def __init__(self, key, data=None):
        self.key = key
        self.data = data
        self.children = {}

class Trie(object):
    def __init__(self):
        self.head = Node(None)
    def insert(self, string):
        curr_node = self.head
        for char in string:
            if char not in curr_node.children:
                curr_node.children[char] = Node(char)
            curr_node = curr_node.children[char]
        curr_node.data = string
            
    def search(self, string):
        curr_node = self.head
        for char in string:
            if char in curr_node.children:
                curr_node = curr_node.children[char]
            else:
                return False
        if curr_node.data != None:
            return True
    def prefix(self,string):
        curr_node = self.head
        prefix_list=[]
        for char in range(len(string)-1):
            if string[char] in curr_node.children:
                curr_node = curr_node.children[string[char]]
                if curr_node.data!=None:
                    prefix_list.append(curr_node.data)
        return prefix_list

t=int(input())
for test in range(t):
    n=int(input())
    index_dict={}
    prefix_dict={}
    trie=Trie()
    
    for i in range(n):
        s=input().strip()
        trie.insert(s)
        index_dict[i]=s
    for i in range(n):
        prefix_list=trie.prefix(index_dict[i])
        if len(prefix_list)==0:
            print(0,end="")
        else:
            flag=False
            for j in prefix_list:
                if trie.search(index_dict[i][len(j):]):
                    flag=True
                    break
            if flag:
                print("1",end="")
            else:
                print("0",end="")
    if test!=t-1:
        print()
```

```python
# D번 트라이 쓸 필요도 없는데 너무 복잡하게 생각한거 아닌가 싶다.... 
for _ in range(int(input())):
    n = int(input())
    s = [input() for _ in range(n)]
    u = set(s)
    ans = []
    for x in s:
        for i in range(1, len(x)):
            if x[:i] in u and x[i:] in u:
                ans.append("1")
                break
        else:
            ans.append("0")
    print("".join(ans))
```

```python
# E Mirror Grid  첫 시도 실패.. 배열 회전 연습 90도 180도 270도
t=int(input())
for _ in range(t):
    n=int(input())
    arr=[list(input().strip()) for _ in range(n)]
    ans=0
    if n%2!=0:
        st=int(n//2)
        for i in range(st):
            for j in range(st):
                cnt_0=0
                if arr[i][j]=='0':
                    cnt_0+=1
                if arr[j][n-1-i]=='0':
                    cnt_0+=1
                if arr[n-1-j][n-1-i]=='0':
                    cnt_0+=1
                if arr[n-1-j][i]=='0':
                    cnt_0+=1
                if cnt_0>=2:
                    ans+=4-cnt_0
                else:
                    ans+=cnt_0
        
        for i in range(st):
            cnt_0=0
            if arr[st][i]=='0':
                cnt_0+=1
            if arr[st][n-1-i]=='0':
                cnt_0+=1
            if arr[i][st]=='0':
                cnt_0+=1
            if arr[n-1-i][st]=='0':
                cnt_0+=1
            if cnt_0>=2:
                ans+=4-cnt_0
            else:
                ans+=cnt_0
        print(ans)
    else:
        st=int(n//2)
        for i in range(st):
            for j in range(st):
                cnt_0=0
                if arr[i][j]=='0':
                    cnt_0+=1
                if arr[j][n-1-i]=='0':
                    cnt_0+=1
                if arr[n-1-j][n-1-i]=='0':
                    cnt_0+=1
                if arr[n-1-j][i]=='0':
                    cnt_0+=1
                if cnt_0>=2:
                    ans+=4-cnt_0
                else:
                    ans+=cnt_0
        print(ans)
```
