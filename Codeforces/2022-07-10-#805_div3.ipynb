```python
# A. Round Down the Price
import sys
input=sys.stdin.readline
t=int(input().strip())
for _ in range(t):
    m=int(input().strip())
    l=len(str(m))
    print(m-10**(l-1))
    
```

```python
# B. Polycarp Writes a String from Memory <더 매끄럽게 짤 수 없나?>
t=int(input()) 
for _ in range(t):
    s=input().strip()
    s_len=len(s)
    days=0
    cnt=0
    while cnt<s_len:
        memory_words=set()
        while len(memory_words)<3:
            if cnt>=s_len:
                break
            if s[cnt] not in memory_words:
                memory_words.add(s[cnt])
            cnt+=1
        if cnt!=s_len:
            while True:
                if cnt>=s_len:
                    break
                if s[cnt] in memory_words:
                    cnt+=1
                else:
                    break
        days+=1
    print(days)
```

```python
# C. Train and Queries
import sys
t=int(input())
for _ in range(t):
    input()
    n,k=map(int,input().split())
    check_index={}
    station=set()
    index=0
    for i in list(map(int,input().split())):
        if i not in station:
            station.add(i)
            check_index.update({i:[index]})
        else:
            check_index[i].append(index)
        index+=1
    for q in range(k):
        a,b=map(int,input().split())
        if a in station and b in station:
            if check_index[a][0]<=check_index[b][-1]:
                print("YES")
            else:
                print("NO")
        else:
            print("NO")
```

```python
# D. Not a Cheap String
t=int(input())
for _ in range(t):
    indice=[[] for _ in range(26)]
    index=0
    w=input().strip()
    for i in w:
        indice[ord(i)-97].append(index)
        index+=1
    check=[0]*index
    p=int(input())
    score=0
    for i in range(26):
        if indice[i]:
            for j in range(len(indice[i])):
                score+=i+1
                if score<=p:
                    check[indice[i][j]]=1
    for i in range(len(w)):
        if check[i]:
            print(w[i],end="")
    print()
    
```

```python
# E. Split Into Two Sets
t=int(input())
for _ in range(t):
    n=int(input())
    dominos=[]
    for _ in range(n):
        a,b=map(int,input().split())
        dominos.append([a,b])
    set1=set()
    set2=set()
    check=True
    for u,v in dominos:
        if u!=v:
            if u not in set1 and v not in set1:
                set1.add(u)
                set1.add(v)
                continue
            if u not in set2 and v not in set2:
                set2.add(u)
                set2.add(v)
                continue
        check=False
    
    if check:
        print("YES")
    else:
        print("NO")
```
